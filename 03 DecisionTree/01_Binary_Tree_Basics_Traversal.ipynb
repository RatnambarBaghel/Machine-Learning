# Binary Tree Node creation


```python
class Node:

  def __init__(self,data):

    self.data = data
    self.left = None
    self.right = None

root = Node("A")

root.left = Node("B")
root.right = Node("C")

root.left.left = Node("D")
root.left.right = Node("E")


```

# Binary Tree Traversal(Pre,Post,In)

```python
# Preorder Root → Left → Right

def preorder(node):
    if node is None:
      return

    print(node.data, end=" ")
    preorder(node.left)
    preorder(node.right)



preorder(root)
```

```python
# Inorder Left -> Root -> Right

def inorder(node):
  if node is None:
    return

  inorder(node.left)

  print(node.data, end =" ")
  inorder(node.right)


inorder(root)
```

```python
# Postorder Left -> Right -> Root

def postorder(node):
  if node is None:
    return

  postorder(node.left)
  postorder(node.right)

  print(node.data, end=" ")

postorder(root)

```

# Binary Tree Algorithms


```python
# Seraching in Binary Tree

def search(node, target):
  if node is None:
    return False

  if node.data == target:
    return True

  if search(node.left, target):
    return True

  return search(node.right, target)


print("Search E",search(root, "E"))
print("Search F",search(root, "F"))
result = search(root, "E")
print(result)
```

```python
# Count in Binary Tree(edge based)

def countHeight(node):
  if node is None:
    return -1

  leftHeight = countHeight(node.left)
  rightHeight = countHeight(node.right)

  return 1 + max(leftHeight, rightHeight)


print("Height of Tree edge based",countHeight(root))

def count_height_node_based(node):
  if node is None:
    return 0

  left_height = count_height_node_based(node.left)
  right_height = count_height_node_based(node.right)

  return 1+ max(left_height , right_height)


print("Height of Tree node based", count_height_node_based(root))


```

```python
#Count Total Nodes in Binary Tree

def count_nodes(node):
  if node is None:
    return 0

  return 1+ count_nodes(node.left) + count_nodes(node.right)

print("Total nodes in a Tree", count_nodes(root))
```

```python
# Count leaf nodes in binary tree

def count_leaf(node):
  if node is None:
    return 0

  if node.left == None and node.right == None:
    return 1

  return count_leaf(node.left) + count_leaf(node.right)


print("Leaf nodes:",count_leaf(root))
```

```python
# Mirror Binary Tree 🪞

def mirror(node):
  if node is None:
    return

  node.left, node.right = node.right, node.left

  mirror(node.left)
  mirror(node.right)


mirror(root)
print(inorder(root))

```
